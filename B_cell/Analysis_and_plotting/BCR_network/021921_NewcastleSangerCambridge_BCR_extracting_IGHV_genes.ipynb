```python
bdata = sc.read_h5ad('h5ad/covid_jan_2021_bcells.h5ad')
bdata = bdata[bdata.obs['Collection_Day'] == 'D0'].copy()
bdata = bdata[bdata.obs['bcr_QC_pass'] == 'True'].copy()
bvdj = ddl.read_h5('dandelion_output/covid_jan_2021_bcells_vdj.h5')
ddl.tl.transfer(bdata, bvdj)
```

```python
test = bdata.obs[['sample_id', 'Status_on_day_collection_summary', 'Sex', 'v_call_heavy']]
test
```

```python
test2 = pd.crosstab(test['sample_id'], test['v_call_heavy']).apply(lambda r: r/r.sum(), axis=1)
s = test2.sum()
test2 = test2[s.sort_values(ascending=False).index[:20]]
test2
```

```python
status_dict = dict(zip(test['sample_id'], test['Status_on_day_collection_summary']))
sex_dict = dict(zip(test['sample_id'], test['Sex']))
```

```python
test2['status'] = [status_dict[x] for x in test2.index]
test2['sex'] = [sex_dict[x] for x in test2.index]
```

```python
test2.to_csv('other_analyses/bdata_obs_for_v_gene_usage.csv')
```
